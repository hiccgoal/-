```python
fp = '/home/jhwang/work/Dataset/user_taggedbookmarks.csv'
```

```python
def loaddata(fp):
    data = [line.strip().split(',')[:3] for line in open(fp).readlines()[1:]]
    return data
```

```python
data = loaddata(fp)
data,len(data)
```

```python
# 计算每个tag及其对应的频率（流行度）
def TagPopularity(records):
    tagfreq=dict()
    for user,item,tag in records:
        if tag not in tagfreq:
            tagfreq[tag] = 0
        tagfreq[tag] += 1
    return tagfreq
```

```python
tagPopularity = TagPopularity(data)
tagPopularity,len(tagPopularity)
```

```python
# 构建一个列表，key为欢迎度，value为对应该欢迎度的标签（不止一个）
from collections import defaultdict
new_tagPopularity = defaultdict(list)
for key, value in tagPopularity.items():
    new_tagPopularity[value].append(key)
new_tagPopularity
```

```python
new_tagPopularity = dict(new_tagPopularity)
```

```python
new_tagPopularity
```

```python
for i in new_tagPopularity:
    new_tagPopularity[i] = len(new_tagPopularity[i])
new_tagPopularity
# 把每个流行度对应的标签list转换为list的长度，即标签个数
```

```python
# 画图
import matplotlib.pylab as plt
lists = sorted(new_tagPopularity.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.show()
```
